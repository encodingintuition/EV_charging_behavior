# EDA
<hr>

```python
# sum of energy per week for 2014

fig = px.histogram(df, x = 'week', y = 'Energy (kWh)',
                color = 'day_of_week', height=400, nbins = 52)

fig.update_layout(title_text='Sum of Energy Used per week for 2014')
fig.update_layout(bargap=0.2)
fig.show()
```

```python
fig_02 = px.histogram(df, x = 'day_of_week', y = 'Energy (kWh)',
                color = 'month',
                category_orders=dict(day_of_week=['Monday','Tuesday','Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']))

fig_02.show()
```

```python
fig_03 = px.histogram(df, x = 'hour', y = 'Energy (kWh)',
                color = 'day_of_week', histfunc='avg', nbins = 24)

fig_03.update_layout(bargap=0.2)
fig_03.show()
```
